```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('/Users/joshuarkarnad/Desktop/Thinkful/Data/NBA Player Data/players_stats.csv')
df['PPG'] = df['PTS'] / df['Games Played']
df['MPG'] = df['MIN'] / df['Games Played']
df.head()
```

```python
Pos_Facet = sns.FacetGrid(df, row="Pos")

Pos_Facet.map(plt.hist, "PPG", color="steelblue",  lw=0)
plt.subplots_adjust(top=0.9)
Pos_Facet.set_axis_labels('PPG by Pos')
plt.suptitle('Plot: PPG by POS')
sns.despine(trim=True)
plt.show()

sns.set(style="whitegrid")

Pos_PPG = sns.boxplot(x='Pos',y='PPG',data=df,palette='deep')  

plt.title('Plot 4: PPG by POS')

sns.despine(offset=10, trim=True)
Pos_PPG.set(xlabel='', ylabel='PPG')

plt.show()

sns.set(style="whitegrid")

g = sns.factorplot(x="Pos", y="PPG", data=df,
                   size=6, kind="bar", palette="deep", ci=95)
g.despine(left=True)
g.set_ylabels("PPG")
g.set_xlabels("Pos")
plt.title('Barplot: PPG by Pos')
plt.show()

sns.set(style="whitegrid")

g = sns.factorplot(x="Pos", y="PPG", data=df,
                   size=6, kind="point", palette="deep",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("PPG")
g.set_xlabels("Pos")
plt.title('Pointplot: PPG by POS')
plt.show()
```

```python
g = sns.lmplot(y='PPG', # Variable 1.
               x='MPG', # Variable 2.
               data=df, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("PPG")
g.set_xlabels("MPG")
plt.title('Scatterplot: PPG by MPG')
plt.show()

g = sns.lmplot(y='PPG', 
               x='MPG',
               data=df, # Data.
               hue='Pos',
               fit_reg=True, 
               ci=False,
               scatter_kws={'alpha':0.4}, 
               palette='deep',
               lowess=True) # Using local weighing to fit a line.
g.set_ylabels("PPG")
g.set_xlabels("MPG")
plt.suptitle('Scatterplot Lowess: PPG by MPG, split by Pos')
plt.subplots_adjust(top=0.9)
plt.show()

g = sns.lmplot(y='PPG', 
               x='MPG',
               data=df, # Data.
               col='Pos',
               fit_reg=True, 
               ci=True,
               scatter_kws={'alpha':0.4}, 
               palette='deep') # Using local weighing to fit a line.
g.set_ylabels("PPG")
g.set_xlabels("MPG")
plt.suptitle('Scatterplot Lowess: PPG by MPG, split by Pos')
plt.subplots_adjust(top=0.9)
plt.show()
```

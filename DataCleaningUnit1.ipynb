```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('/Users/joshuarkarnad/Desktop/Thinkful/Data/WELLCOME/WELLCOME_APCspend2013_forThinkful.csv', encoding='iso-8859-1')
```

```python
df = df.dropna()
df.columns = ['ID', 'Publisher', 'Journal','Article', 'Cost']
df['Journal'] = df['Journal'].str.lower()
df['Journal'] = df['Journal'].str.replace(' ', '')
df['Journal'] = df['Journal'].str.replace(',', '')
df['Journal'] = df['Journal'].str.replace(':', '')
df['Cost'] = df['Cost'].str.replace('Â£', '')
df['Cost'] = df['Cost'].str.replace('$', '')
df['Cost'] = df['Cost'].astype(float)

df.head()
```
